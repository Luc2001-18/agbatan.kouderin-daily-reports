### Exercise 8:
Add titles, axis labels, and a legend to the plot in Exercise 7.

```python
# Exercise 8
```

````python
x=np.linspace(0,10, 100)
y1=x**2
y2=x**3
plt.plot(x, y1, label='$y=x^2$')
plt.plot(x,y2, label='$y=x^3$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('PLot of $x^2$ and $x^3$')
plt.legend(loc=2)
plt.show()
````

### Exercise 9:
Create a subplot with 2 rows and 1 column. Plot `y1 = x^2` in the first subplot and `y2 = x^3` in the second.

```python
# Exercise 9
```

````python
x=np.linspace(0,10, 100)

fig, axes = plt.subplots(2,1, figsize=(4,8))
y1=x**2
y2=x**3
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].plot(x, y1, label='$y=x^2$')
axes[0].set_title('PLot of $x^2$')
axes[0].legend(loc=2)

axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].plot(x,y2, label='$y=x^3$')
axes[1].set_title('PLot of $x^2$ and $x^3$')
axes[1].legend(loc=2)
fig.tight_layout()
plt.show()
````

### Exercise 10:
Use a different color and linestyle for each line in a plot of three functions: `y1 = x`, `y2 = x^2`, and `y3 = x^3`.

```python
# Exercise 10
```

````python
help(plt.plot)
````

````python
x=np.linspace(0,10, 100)
fig, axes = plt.subplots(1,2, figsize=(14,8))

y3=x
y1=x**2
y2=x**3
axes[0].plot(x, y3, label='$y=x$', linestyle= 'dashed')
axes[0].plot(x, y1, label='$y=x^2$', linestyle='dashdot')
#axes[0].set_yscale("log")
axes[0].plot(x,y2, label='$y=x^3$', linestyle='dotted')
#axes[0].xlabel('x')
#axes[0].ylabel('y')
axes[0].set_title('PLot of $x^2$ and $x^3$')
axes[0].legend(loc=2)
plt.show()
````

````python
x=np.linspace(-10,10, 100)
fig, axes = plt.subplots(1,3, figsize=(14,8))

y3=x
y1=x**2
y2=x**3
axes[0].plot(x, y3, label='$y=x$', linestyle= 'dashed')
axes[1].plot(x, y1, label='$y=x^2$', linestyle='dashdot')
#axes[0].set_yscale("log")
axes[2].plot(x,y2, label='$y=x^3$', linestyle='dotted')
#axes[0].xlabel('x')
#axes[0].ylabel('y')
axes[0].set_title('PLot of $x$')
axes[1].set_title('PLot of $x^2$')
axes[2].set_title('PLot of $x^3$')
axes[0].legend(loc=2)
axes[1].legend(loc=2)
axes[2].legend(loc=2)
plt.show()
````
